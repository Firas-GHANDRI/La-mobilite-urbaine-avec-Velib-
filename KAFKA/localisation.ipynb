```python
import requests
import pandas as pd
import json
```

```python
response=requests.get("https://velib-metropole-opendata.smovengo.cloud/opendata/Velib_Metropole/station_information.json")
```

```python
data=response.content
type(data)
```

```python
# D'abord, on décode les bytes en string
json_string =data.decode('utf-8')

# Ensuite, on transforme le string en JSON
data = json.loads(json_string)
```

```python
data["data"]["stations"][0]
```

```python
longueur=len(data["data"]["stations"])
longueur
```

```python
station_id_list=[]
for i in range(0,longueur):
    station_id_list.append(data["data"]["stations"][i]["station_id"])

station_id_list
```

```python
name_list=[]
for i in range(0,longueur):
    name_list.append(data["data"]["stations"][i]["name"])

name_list
```

```python
lat_list=[]
for i in range(0,longueur):
    lat_list.append(data["data"]["stations"][i]["lat"])

lat_list
```

```python
lon_list=[]
for i in range(0,longueur):
    lon_list.append(data["data"]["stations"][i]["lon"])

lon_list
```

```python
capacity_list=[]
for i in range(0,longueur):
    capacity_list.append(data["data"]["stations"][i]["capacity"])

capacity_list
```

```python
df=pd.DataFrame(list(zip(station_id_list,name_list,lat_list,lon_list,capacity_list)),columns=["id","name","lat","lon","capacity"])
df.head(40)
```

```python
df.to_csv("localisation.csv")
```
